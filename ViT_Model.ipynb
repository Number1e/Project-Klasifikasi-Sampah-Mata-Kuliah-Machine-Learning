```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Layer, Dense, Dropout, LayerNormalization,
    MultiHeadAttention, Embedding, Input,
    GlobalAveragePooling1D, Add, Rescaling
)
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt
import numpy as np
import os

# ==========================================
# 1. KONFIGURASI "ViT-MINI" (ANTI OOM)
# ==========================================
# Kita mengecilkan ukuran model agar muat di GPU Colab & cocok untuk dataset kecil

# Ukuran Input & Batch
UKURAN_GAMBAR = (128, 128) # Diturunkan dari 224 ke 128 untuk hemat memori
BATCH_SIZE = 32

# Parameter Arsitektur ViT (Disederhanakan)
PATCH_SIZE = 16
NUM_PATCHES = (UKURAN_GAMBAR[0] // PATCH_SIZE) ** 2 # (128/16)^2 = 64 Patches
PROJECTION_DIM = 64        # Diturunkan drastis dari 768 ke 64
NUM_HEADS = 4              # Diturunkan dari 12 ke 4
TRANSFORMER_LAYERS = 4     # Diturunkan dari 12 ke 4
MLP_UNITS = [PROJECTION_DIM * 2, PROJECTION_DIM]
MLP_DROPOUT = 0.1

# Path & Label
PATH_TRAINING = '/content/drive/MyDrive/Machine Learning/Dataset Sampah/training'
PATH_VALIDATION = '/content/drive/MyDrive/Machine Learning/Dataset Sampah/validation'
LABEL_KELAS = ["Kaca", "Kardus", "Plastik"]
JUMLAH_KELAS = len(LABEL_KELAS)

print(f"Konfigurasi: Gambar {UKURAN_GAMBAR}, Dimensi {PROJECTION_DIM}, Layer {TRANSFORMER_LAYERS}")

# ==========================================
# 2. DEFINISI LAPISAN KUSTOM ViT
# ==========================================

class Patches(Layer):
    def __init__(self, patch_size, **kwargs):
        super(Patches, self).__init__(**kwargs)
        self.patch_size = patch_size

    def call(self, images):
        batch_size = tf.shape(images)[0]
        patches = tf.image.extract_patches(
            images=images,
            sizes=[1, self.patch_size, self.patch_size, 1],
            strides=[1, self.patch_size, self.patch_size, 1],
            rates=[1, 1, 1, 1],
            padding="VALID",
        )
        patch_dims = patches.shape[-1]
        patches = tf.reshape(patches, [batch_size, -1, patch_dims])
        return patches

    def get_config(self):
        config = super(Patches, self).get_config()
        config.update({"patch_size": self.patch_size})
        return config

class PatchEncoder(Layer):
    def __init__(self, num_patches, projection_dim, **kwargs):
        super(PatchEncoder, self).__init__(**kwargs)
        self.num_patches = num_patches
        self.projection_dim = projection_dim
        self.projection = Dense(units=projection_dim)
        self.position_embedding = Embedding(
            input_dim=num_patches, output_dim=projection_dim
        )

    def call(self, patch):
        positions = tf.range(start=0, limit=self.num_patches, delta=1)
        encoded = self.projection(patch) + self.position_embedding(positions)
        return encoded

    def get_config(self):
        config = super(PatchEncoder, self).get_config()
        config.update({"num_patches": self.num_patches, "projection_dim": self.projection_dim})
        return config

def mlp(x, hidden_units, dropout_rate):
    for units in hidden_units:
        x = Dense(units, activation=tf.nn.gelu)(x)
        x = Dropout(dropout_rate)(x)
    return x

# ==========================================
# 3. DATA GENERATOR
# ==========================================

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,      # Augmentasi sedikit dikurangi agar stabil
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

validation_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    PATH_TRAINING,
    target_size=UKURAN_GAMBAR, # Menggunakan 128x128
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

validation_generator = validation_datagen.flow_from_directory(
    PATH_VALIDATION,
    target_size=UKURAN_GAMBAR, # Menggunakan 128x128
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

# ==========================================
# 4. BUILD MODEL
# ==========================================

def create_mini_vit():
    inputs = Input(shape=(UKURAN_GAMBAR[0], UKURAN_GAMBAR[1], 3))

    # Patching
    patches = Patches(PATCH_SIZE)(inputs)
    encoded_patches = PatchEncoder(NUM_PATCHES, PROJECTION_DIM)(patches)

    # Transformer Blocks
    for _ in range(TRANSFORMER_LAYERS):
        # Layer Norm 1
        x1 = LayerNormalization(epsilon=1e-6)(encoded_patches)
        # Attention
        attention_output = MultiHeadAttention(
            num_heads=NUM_HEADS, key_dim=PROJECTION_DIM, dropout=MLP_DROPOUT
        )(x1, x1)
        # Skip Connection 1
        x2 = Add()([attention_output, encoded_patches])

        # Layer Norm 2
        x3 = LayerNormalization(epsilon=1e-6)(x2)
        # MLP
        x3 = mlp(x3, hidden_units=MLP_UNITS, dropout_rate=MLP_DROPOUT)
        # Skip Connection 2
        encoded_patches = Add()([x3, x2])

    # Output Head
    representation = LayerNormalization(epsilon=1e-6)(encoded_patches)
    representation = GlobalAveragePooling1D()(representation)

    # Classifier sederhana
    features = Dense(256, activation='relu')(representation) # Hidden layer dikurangi
    features = Dropout(0.5)(features)
    logits = Dense(JUMLAH_KELAS, activation='softmax')(features)

    model = Model(inputs=inputs, outputs=logits)
    return model

# Buat Model
model = create_mini_vit()

# Compile
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), # LR standar bisa dipakai untuk model kecil
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ==========================================
# 5. TRAINING
# ==========================================

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = ModelCheckpoint('model_vit_mini.keras', monitor='val_accuracy', save_best_only=True)

print("\nMulai melatih ViT Mini...")
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // BATCH_SIZE,
    epochs=50, # Epoch diperbanyak karena model dilatih dari nol
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // BATCH_SIZE,
    callbacks=[early_stop, checkpoint]
)

# ==========================================
# 6. EVALUASI
# ==========================================

# Plot
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss')
plt.legend()
plt.show()

# Hasil Akhir
loss, accuracy = model.evaluate(validation_generator)
print(f"\nAkurasi Validasi Akhir: {accuracy*100:.2f}%")
# ==========================================
# 7. EVALUASI MENDALAM (CONFUSION MATRIX & REPORT)
# ==========================================
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import pandas as pd

# 1. Ambil Prediksi
# Pastikan generator di-reset agar mulai dari awal
validation_generator.reset()

print("\nMenghasilkan prediksi untuk data validasi...")
# predict() menghasilkan probabilitas (softmax output)
Y_pred_prob = model.predict(validation_generator)

# Ambil kelas dengan probabilitas tertinggi (argmax)
Y_pred_classes = np.argmax(Y_pred_prob, axis=1)

# Ambil label asli (Ground Truth) dari generator
Y_true = validation_generator.classes

# Pastikan mapping label sesuai urutan generator (alfabetis biasanya)
# class_indices berupa {'Kaca': 0, 'Kardus': 1, ...}
class_labels = list(validation_generator.class_indices.keys())

# -------------------------------------------------------
# A. CONFUSION MATRIX (Visualisasi Heatmap)
# -------------------------------------------------------
cm = confusion_matrix(Y_true, Y_pred_classes)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_labels,
            yticklabels=class_labels)
plt.title('Confusion Matrix: Prediksi vs Aktual')
plt.ylabel('Label Asli (Ground Truth)')
plt.xlabel('Label Prediksi Model')
plt.show()

# -------------------------------------------------------
# B. TABEL EVALUASI (Classification Report)
# -------------------------------------------------------
print("\n" + "="*40)
print("TABEL EVALUASI MODEL (CLASSIFICATION REPORT)")
print("="*40)

# Membuat report dalam bentuk dictionary agar bisa jadi DataFrame pandas
report_dict = classification_report(Y_true, Y_pred_classes, target_names=class_labels, output_dict=True)
df_report = pd.DataFrame(report_dict).transpose()

# Format tampilan agar rapi
print(df_report)

# Insight Tekstual Sederhana
print("\n" + "="*40)
print("INSIGHT SINGKAT:")
for label in class_labels:
    f1 = report_dict[label]['f1-score']
    if f1 < 0.6:
        print(f"⚠️ Performa buruk pada kelas '{label}' (F1: {f1:.2f}). Cek apakah sampel kurang atau mirip kelas lain.")
    elif f1 > 0.85:
        print(f"✅ Model sangat baik mengenali '{label}' (F1: {f1:.2f}).")
```
