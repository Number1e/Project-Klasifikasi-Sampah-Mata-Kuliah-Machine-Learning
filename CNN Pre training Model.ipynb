```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.applications import MobileNetV2 # Menggunakan Transfer Learning
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt
import numpy as np
import os

# --- 1. KONFIGURASI ---
# Ganti dengan path Anda
PATH_TRAINING = '/content/drive/MyDrive/Machine Learning/Dataset Sampah/training'
PATH_VALIDATION = '/content/drive/MyDrive/Machine Learning/Dataset Sampah/validation'

# Label akan diambil otomatis dari nama folder oleh Keras
LABEL_KELAS = ["Kaca", "Kardus", "Plastik"] # Hanya untuk referensi plot
JUMLAH_KELAS = len(LABEL_KELAS)

# MobileNetV2 bekerja baik dengan ukuran ini
UKURAN_GAMBAR = (224, 224)
BATCH_SIZE = 32

# --- 2. MEMPERSIAPKAN DATA (DATA GENERATORS) ---
# Ini adalah cara Keras memuat gambar sekaligus melakukan augmentasi

# Augmentasi untuk data training
train_datagen = ImageDataGenerator(
    rescale=1./255,         # Normalisasi piksel dari [0, 255] ke [0, 1]
    rotation_range=30,      # Augmentasi: putar gambar
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,   # Augmentasi: balik horizontal
    fill_mode='nearest'
)

# TIDAK ada augmentasi untuk data validasi. Hanya normalisasi.
validation_datagen = ImageDataGenerator(rescale=1./255)

# Buat generator
train_generator = train_datagen.flow_from_directory(
    PATH_TRAINING,
    target_size=UKURAN_GAMBAR,
    batch_size=BATCH_SIZE,
    class_mode='categorical' # karena kita punya > 2 kelas
)

validation_generator = validation_datagen.flow_from_directory(
    PATH_VALIDATION,
    target_size=UKURAN_GAMBAR,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False # Tidak perlu di-shuffle untuk validasi
)

print(f"Kelas ditemukan: {train_generator.class_indices}")

# --- 3. MEMBANGUN ARSITEKTUR MODEL (TRANSFER LEARNING) ---

# Muat model dasar (MobileNetV2) tanpa lapisan klasifikasi atas
# 'imagenet' berarti kita menggunakan bobot (weights) yang sudah dilatih
base_model = MobileNetV2(weights='imagenet', include_top=False,
                         input_shape=(UKURAN_GAMBAR[0], UKURAN_GAMBAR[1], 3))

# Bekukan (freeze) lapisan-lapisan dari model dasar
# Kita tidak ingin melatih ulang apa yang sudah dipelajari MobileNetV2
base_model.trainable = False

# Bangun model kita di atas model dasar
model = Sequential([
    base_model, # Model dasar

    # Lapisan kustom kita
    GlobalAveragePooling2D(), # Meratakan output fitur dari base_model
    Dense(256, activation='relu'), # Lapisan tersembunyi
    Dropout(0.5),                  # Regularisasi untuk mencegah overfitting
    Dense(JUMLAH_KELAS, activation='softmax') # Lapisan output (3 neuron, 1 per kelas)
])

# --- 4. KOMPILASI MODEL ---
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy', # Standar untuk klasifikasi multi-kelas
    metrics=['accuracy']
)

model.summary() # Tampilkan arsitektur model

# --- 5. MELATIH MODEL ---

# Callbacks untuk menghentikan jika tidak ada kemajuan dan menyimpan model terbaik
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
checkpoint = ModelCheckpoint('model_sampah_terbaik.h5', monitor='val_accuracy', save_best_only=True)

print("Mulai melatih model CNN...")

history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // BATCH_SIZE,
    epochs=30, # Jumlah epoch (bisa ditambah jika perlu)
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // BATCH_SIZE,
    callbacks=[early_stop, checkpoint]
)

print("Pelatihan CNN selesai.")

# --- 6. EVALUASI MODEL ---

# Plot learning curves
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title('Accuracy vs. Epochs')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title('Loss vs. Epochs')
plt.show()

# Evaluasi akhir pada data validasi
loss, accuracy = model.evaluate(validation_generator)
print(f"Akurasi Validasi Akhir: {accuracy*100:.2f}%")

# (Optional) Membuat Confusion Matrix untuk CNN
# y_pred = np.argmax(model.predict(validation_generator), axis=1)
# y_true = validation_generator.classes
# ... (prosesnya mirip dengan SVM, tapi sedikit lebih kompleks karena generator)
```

```python
# --- 7. DEBUGGING MODEL & ERROR ANALYSIS ---
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.image as mpimg

# A. Pastikan generator di-reset agar urutan prediksi sesuai dengan urutan file
validation_generator.reset()

# B. Dapatkan prediksi probabilitas untuk seluruh data validasi
print("Sedang menghasilkan prediksi untuk data validasi...")
Y_pred = model.predict(validation_generator)
y_pred_indices = np.argmax(Y_pred, axis=1) # Konversi probabilitas ke index kelas (0, 1, 2)

# C. Ambil label asli (Ground Truth)
y_true_indices = validation_generator.classes

# D. Ambil nama file dan label kelas
filenames = validation_generator.filenames
class_labels = list(validation_generator.class_indices.keys())

# --- 8. VISUALISASI CONFUSION MATRIX ---
print("\n--- Confusion Matrix ---")
cm = confusion_matrix(y_true_indices, y_pred_indices)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_labels, yticklabels=class_labels)
plt.ylabel('Label Sebenarnya (Actual)')
plt.xlabel('Label Prediksi (Predicted)')
plt.title('Confusion Matrix: Dimana Model Anda Bingung?')
plt.show()

# Tampilkan laporan klasifikasi teks
print("\n--- Classification Report ---")
print(classification_report(y_true_indices, y_pred_indices, target_names=class_labels))

# --- 9. GALERI KESALAHAN (WRONG PREDICTIONS GALLERY) ---
# Ini adalah bagian inti untuk melihat gambar yang salah
print("\n--- Visualisasi Gambar yang Salah Terklasifikasi ---")

# Cari indeks dimana prediksi TIDAK SAMA dengan label asli
errors = np.where(y_pred_indices != y_true_indices)[0]
print(f"Total kesalahan ditemukan: {len(errors)} dari {len(y_true_indices)} gambar.")

# Batasi tampilan agar tidak terlalu banyak (misal: maksimal 15 gambar)
max_display = 15
display_indices = errors[:max_display]

if len(errors) > 0:
    num_rows = (len(display_indices) // 3) + 1
    plt.figure(figsize=(15, 5 * num_rows))

    for i, error_idx in enumerate(display_indices):
        # Path file gambar
        file_path = os.path.join(PATH_VALIDATION, filenames[error_idx])

        # Load gambar asli (bukan tensor yang sudah di-rescale, agar lebih enak dilihat mata)
        img = mpimg.imread(file_path)

        # Info kelas
        true_label = class_labels[y_true_indices[error_idx]]
        pred_label = class_labels[y_pred_indices[error_idx]]
        confidence = Y_pred[error_idx][y_pred_indices[error_idx]] # Seberapa yakin modelnya?

        plt.subplot(num_rows, 3, i + 1)
        plt.imshow(img)
        plt.axis('off')

        # Judul merah menandakan kesalahan
        plt.title(f"Asli: {true_label}\nPrediksi: {pred_label}\nConf: {confidence:.2f}", color='red')

    plt.tight_layout()
    plt.show()
else:
    print("Hebat! Model memprediksi semua data validasi dengan benar.")
```
